```python
# problem 3

def inttoStr(i):
    digit = "012346789"
    if i ==0:
        return "0"
    result = ""
    while i>0:
        result = digit[i%10] + result
        i = i//10
    return result


print(inttoStr(12345))

# Always use python tutor  for understanding the code

# last watched video feb 4 2026 9pm video -- 1:02:17

```
